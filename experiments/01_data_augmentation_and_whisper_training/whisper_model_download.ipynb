**목적**: HuggingFace Hub에 업로드된 파인튜닝 완료 모델(`haseong8012/whisper-tiny_child-50k`)을 로컬로 다운로드하여 저장.

**내용**: 모델 및 프로세서를 `from_pretrained()`으로 로드한 뒤 `save_pretrained()`로 로컬 체크포인트 저장.


```python
model_path = '/content/drive/MyDrive/JBNU/2024-2/Capstone/Whisper/Models/whisper-tiny_child'
```

```python
model_name = 'haseong8012/whisper-tiny_child-50k'
```

```python
from transformers import WhisperForConditionalGeneration
from transformers import WhisperProcessor
```

```python
# 모델 불러오기
model = WhisperForConditionalGeneration.from_pretrained(model_name)
# 프로세서 불러오기
processor = WhisperProcessor.from_pretrained(model_name)

model.eval()
```

```python
model.save_pretrained(model_path)
processor.save_pretrained(model_path)
```
